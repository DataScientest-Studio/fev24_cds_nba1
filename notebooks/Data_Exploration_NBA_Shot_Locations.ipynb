**Distribution des tirs en fonction des années**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les données
df = pd.read_csv('NBA_Shot_Locations_1997to2020.csv')

# Convertir 'Game Date' en datetime
df['Game Date'] = df['Game Date'].astype(str)
df['Year'] = df['Game Date'].str[:4].astype(int)

# Filtrer les années
df = df[(df['Year'] >= 1997) & (df['Year'] <= 2020)]

# Créer une grille de sous-graphiques pour chaque année
years = df['Year'].unique()
num_years = len(years)
num_cols = 4  # Nombre de colonnes pour la grille (vous pouvez ajuster selon le nombre d'années)
num_rows = (num_years + num_cols - 1) // num_cols  # Calculer le nombre de lignes nécessaire

# Créer la grille de sous-graphiques
fig, axes = plt.subplots(num_rows, num_cols, figsize=(16, 12), sharex=True, sharey=True)

# Parcourir chaque année et tracer la distribution de location X par rapport à la location Y
for i, year in enumerate(years):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]
    
    data_year = df[df['Year'] == year]
    sns.scatterplot(data=data_year, x='X Location', y='Y Location', ax=ax)
    ax.set_title(f'Distribution des tirs en {year}')
    ax.set_xlabel('Location X')
    ax.set_ylabel('Location Y')

# Supprimer les sous-graphiques non utilisés
for i in range(len(years), num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    fig.delaxes(axes[row, col])

plt.tight_layout()
plt.show()
```
